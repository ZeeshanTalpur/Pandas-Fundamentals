**Dataset**

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "Date":[
        "2025-01-01",
        "2025-01-02",
        "2025-01-03",
        "2025-01-04",
        "2025-01-05"
    ],
    "Sales":[
        100,
        120,
        110,
        150,
        130
    ]
})
df
```

**1. Converting Strings to DateTime**

```python
print(df["Date"].dtype)
```

This implies these are just strings\
**Solution**

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
print(df["Date"].dtype)
```

Now, Python understands Years, Months, Days, Hours, Minutes, Seconds

**2. The .dt Accessor**\
Once a column becomes DateTime:

```python
df["Date"].dt
```

unlocks many features.

Extract Year:

```python
df["Year"] = df["Date"].dt.year
```

```python
df
```

Extract Month:

```python
df["Month"] = df["Date"].dt.month
```

```python
df
```

Extract Day:

```python
df["Day"] = df["Date"].dt.day
```

```python
df
```

Extract month name:

```python
df["Month Name"] = df["Date"].dt.month_name()
```

```python
df
```

```python
df["Weekday Name"] = df["Date"].dt.day_name()
```

```python
df
```

**6. Date Ranges**

Generate dates:

```python
pd.date_range(
    start = "2025-01-02",
    periods = 5
)
```

Custom Frequency

Monthly:

```python
pd.date_range(
    start = "2025-02",
    periods = 6,
    freq = "ME"
)
```

Common Frequencies\
Code	Meaning\
D-->Daily\
W-->Weekly\
M-->Month End\
MS-->Month Start\
Q-->Quarterly\
Y-->Year End\
H-->Hourly\
min-->Minutes\
S-->Seconds

**7. Resampling**\
This is a groupby for time

Monthly Sales:

```python
print(df.resample(
    "ME"
).sum())
```

Example:

```python
dates = pd.date_range(
    "2025-01-01",
    periods=60
)
sales = np.random.randint(
    100,
    200,
    60
)
df = pd.DataFrame({
    "Sales":sales
},index = dates)
```

```python
df
```

Monthly Totals:

```python
df.resample(
    "ME"
).sum()
```

Monthly Averages:

```python
df.resample(
    "ME"
).mean()
```

Intuition:\
groupby() --> categories\
resample() --> Time periods

**8. Rolling Windows**

Question: Average sales over the last 3 days?

```python
df["Rolling_3"] = df["Sales"].rolling(3).mean()
```

```python
df
```

Intuition:\
Takes\
current day + previous 2 days = 3 days in total\
Average\
This is called:\
Moving Average\
Because it shows how the average of immediate 3 days moves or changes with every single day

**9. Expanding Windows**\
rolling(n) --> Last n values\
expanding() --> Everything so far

```python
df["Running Average"] = df[
    "Sales"
    ].expanding().mean()
```

```python
df
```

**10. Shifting**

```python
df["Yesterday"] = df[
    "Sales"
    ].shift(1)
```

```python
df
```

Difference between Days:

```python
df["Difference"] = df[
    "Sales"
    ].diff()
```

```python
df
```

Percentage Growth:

```python
df["Growth"] = df["Sales"].pct_change()
```

```python
df
```
