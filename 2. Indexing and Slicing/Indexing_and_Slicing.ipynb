# <center> Indexing and Slicing Mastery

In Pandas, we still do selection like NumPy, but now we have:\
Labels\
Column Names\
Indexes\
Positions\
Conditions\
This makes Pandas much more expressive.

**The Big Idea**\
A DataFrame has:\
Rows\
Columns\
You must learn to access both.\
**Think:**\
Which rows?\
Which columns?\
Every selection problem is some combination of those two questions.

Let's create our dataset to work with:

```python
import pandas as pd
df = pd.DataFrame({
    "Name":["Ali","Sara","Ahmed","Ayesha","Bilal"],
    "Age":[22,25,30,28,35],
    "Salary":[50000,70000,90000,80000,120000],
    "Department":["HR","IT","IT","Finance","HR"]
})
df.index=[101,102,103,104,105]
print(df)
```

**1. Selecting Columns**\
Columns are selected using names.\
**Single Column**

```python
df["Salary"]
```

**Multiple Columns**

```python
df[["Name","Salary"]]
```

**2. Row Selection**\
Now we select rows.\
Pandas gives us:\
.loc[]\
.iloc[]\
These are extremely important.

**loc[]**\
Think: LABEL BASED

```python
df.loc[102]
```

Multiple rows:

```python
df.loc[[101,104]]
```

Range:

```python
df.loc[101:103]
```

Notice: Upper limit is also inclusive when working with loc[], unlike NumPy Slicing.

**iloc[]**\
Think: INTEGER POSITION BASED

```python
df.iloc[1]
```

Multiple rows:

```python
df.iloc[[1,4]]
```

Range:

```python
df.iloc[2:4]
```

Notice: iloc[] doesn't include upper limit, works exactly like NumPy slicing and unlike loc[].

**Selecting Rows and Columns Together**\
**1. loc**\
Syntax: df.loc[row_selection, column_selection]


```python
df.loc[103,"Department"]
```

```python
df.loc[102:105,["Age","Salary"]]
```

**2. iloc**\
Syntax: df.iloc[row_selection,column_selection]

```python
df.iloc[3,1]
```

```python
df.iloc[1:4,[0,2,3]]
```

**3. Fast Scalar Access**\
When accessing one value.\
**a. at[]**\
Label based.

```python
df.at[102,"Department"]
```

**b. iat[]**\
Position based.

```python
df.iat[1,0]
```

These are faster than:\
df.loc[]\
df.iloc[]\
for single values.

**4. Boolean Filtering**\
This is where Pandas becomes powerful.\
**Question:**\
Show employees with salary > 70000

```python
df[df["Salary"]>70000]
```

**Multiple Conditions:**

AND:

```python
df[
    (df["Salary"]>70000)
    &
    (df["Department"]=="IT")
    ]
```

OR:

```python
df[
    (df["Salary"]>70000)
    |
    (df["Department"]=="IT")
    ]
```

NOT:

```python
df[~(df["Department"]=="IT")]
```

**5. Query Method**\
Same filtering.\
More SQL-like.

```python
df.query("Salary>65000")
```

```python
df.query("Salary>70000 and Department=='IT'")
```

**6. Sorting:**

Ascending:

```python
df.sort_values("Salary")
```

Descending:

```python
df.sort_values("Salary",ascending=False)
```

Multiple columns:

```python
df.sort_values(["Salary","Age"])
```

**7. Membership Testing**\
Very common.

```python
df[df["Department"].isin(["HR","IT"])]
```

**8. Unique Values**\
Often used before analysis.

```python
df["Department"].unique()
```

```python
df["Department"].nunique()
```
