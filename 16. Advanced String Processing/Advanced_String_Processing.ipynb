# <center>Advanced String Processing

If Window Functions are the king of numerical analytics, String Processing is the king of data cleaning.\
But Most of the real-world datasets are not messy because of numbers but because of **text**\
Before you build a model, you'll spend a lot of time cleaning these values

**Dataset**

```python
import pandas as pd
df = pd.DataFrame({
    "Name":[
        " ali ",
        "SARA",
        "ahMed",
        " John Doe "
    ],
    "Email":[
        "ali@gmail.com",
        "SARA@Yahoo.com",
        "ahmed123@gmail.com",
        "john@company.org"
    ],
    "Department":[
        "IT Department",
        "HR Department",
        "Finance Department",
        "IT Department"
    ],
    "Code":[
        "EMP-001",
        "EMP-002",
        "EMP-003",
        "EMP-004"
    ]
})
df
```

**Important** : Strings in Pandas are manipulated through the **.str** accessor

**1. lower()**\
converts everything to lowercase

```python
df["Name"].str.lower()
```

**2. upper()**\
converts everything to uppercase

```python
df["Name"].str.upper()
```

**3. title()**\
converts everything to title case

```python
df["Name"].str.title()
```

**4. strip()**\
Removes leading and trailing spaces

```python
df["Name"].str.strip()
```

There are also:\
lstrip() --> Removes leading spaces\
rstrip() --> Removes trailig spaces

**5. replace()**\
Replace text

```python
df["Department"].str.replace(
    "Department",
    ""
)
```

Regex replacement:

```python
df["Department"].str.replace(
    r"\d",
    "",
    regex=True
)
```

It removes every digit

**6. contains()**\
Find text

```python
df["Email"].str.contains(
    "gmail"
)
```

Case insensitive check:

```python
df["Email"].str.contains(
    "gmail",
    case=False
)
```

Regex:\
Eg: Contains numbers?

```python
df["Email"].str.contains(
    r"\d"
)
```

**7. startswith() and endswith()**

```python
df["Code"].str.startswith("EMP")
```

```python
df["Email"].str.endswith(".com")
```

**8. split()**\
breaks a string

```python
df["Name"].str.split()
```

Split and Exapnd into columns:

```python
df["Name"].str.split(
    expand=True
)
```

**9. cat()**\
Concatenate strings

```python
df["Name"].str.cat(
    df["Department"],
    sep = '-'
)
```

**10. len()**

```python
df["Email"].str.len()
```

**11. slice()**

```python
df["Code"].str.slice(2)
```

Equivalent to: .str[4: ]

**12. extract()**\
Extract digits:

```python
df["Code"].str.extract(
    r"(\d+)"
)
```

Extract letters:

```python
df["Code"].str.extract(
    r"([A-Z]+)"
)
```

Extract Email Username:

```python
df["Email"].str.extract(
    r"(^.*)@"
)
```

**13. extractall()**

```python
s = pd.Series(["A12B45C89"])

s.str.extractall(
    r"(\d+)"
)
```

Unlike extract(), it finds every single match

**14. findall()**

```python
s.str.findall(
    r"(\d+)"
)
```

**15. match()**\
Check whether the beginning matches

```python
df["Code"].str.match(
    r"EMP"
)
```

**fullmatch():** Whole String match

```python
df["Code"].str.fullmatch(
    r"EMP-\d{3}"
)
```
