# <center>GroupBy Mastery

**This is arguably the most important Pandas phase.**\
Many people know Pandas.\
Far fewer truly understand groupby().\
Yet, most business questions are basically:\
Take a group of people/items and calculate something about them.\
**Examples:**\
Average salary per department\
Total sales per city\
Maximum spending per customer type\
Average marks per class\
Number of orders per month\
All of these are GroupBy problems.\
**The Core Philosophy**\
Pandas follows: Split --> Apply --> Combine

**Dataset**

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["Ali","Sara","Ahmed","Ayesha","Bilal","Usman"],
    "Department":["HR","IT","IT","Finance","HR","Finance"],
    "Age":[22,25,30,28,35,40],
    "Salary":[50000,70000,90000,80000,120000,100000]
})
df
```

**1. Basic GroupBy**\
This one doesn't compute anything.\
It just creates groups.

```python
group = df.groupby("Department")
print(group)
```

```python
group.groups
```

```python
group.groups.keys()
```

```python
list(group.groups.keys())
```

This is just preparing the data for the group operations

**2. Mean**

```python
df.groupby("Department")["Salary"].mean()
```

Interpretation : Average Salary per Department

**3. Sum**

```python
df.groupby("Department")["Salary"].sum()
```

Interpretation : Total Salary per Department

**4. Count**

```python
df.groupby("Department")["Name"].count()
```

Alternative: size()

```python
df.groupby("Department")["Name"].size()
```

**Difference:**\
count() ignores NaN\
size() counts everything

**5. Maximum**

```python
df.groupby("Department")["Salary"].max()
```

Interpretation : Highest Salary per Department

**6. Minimum**

```python
df.groupby("Department")["Salary"].min()
```

Interpretation : Lowest Salary per Department

**7. Median**

```python
df.groupby("Department")["Salary"].median()
```

**8. Standard Deviation**

```python
df.groupby("Department")["Salary"].std()
```

Interpretation:\
Large Std : Employees earn very different salaries.\
Small Std : Employees earn similar salaries.

**9. Multiple Aggregations**\
Real analysts rarely compute only one thing.

```python
df.groupby("Department")["Salary"].agg(
    ["sum","mean","median","max","min","count","std"]
)
```

**10. Named Aggregations**\
Professionals often rename outputs.

```python
df.groupby("Department").agg(
    AverageSalary = ("Salary","mean"),
    HighestSalary = ("Salary","max"),
    LowestSalary = ("Salary","min"),
    EmployeeCount = ("Name","size")
)
```

**11. Grouping By Multiple Columns**

```python
Gender = pd.Series(["Male","Female","Male","Female","Male","Male"])
df["Gender"]=Gender
```

```python
df
```

```python
df.groupby(
    ["Department","Gender"]
)["Salary"].mean()
```

This is called a Multi Index

**12. as_index = False**

```python
df.groupby(
    "Department",
    as_index=False
)["Salary"].mean()
```

Notice that with as_index = False, the department becomes a normal column rather than the index itself. It is useful before exporting to CSV.

**13. Sorting Group Results**

```python
df.groupby("Department")["Salary"].mean().sort_values(ascending=False)
```
