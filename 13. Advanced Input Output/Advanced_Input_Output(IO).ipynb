# <center>Advanced Input/Output (I/O)

**1. CSV Files**\
CSV(Comma-Separated Values) is by far the most common data format in Data Science.

**Reading a CSV**

```python
import pandas as pd
df = pd.read_csv("customers.csv")
print(df)
```

read_csv() is probably the single most-used Pandas function in industry

**Saving a CSV**

```python
df.to_csv("output.csv")
```

This creates an extra column because Pandas always saves the index

Correct way:

```python
df.to_csv(
    "output.csv",
    index=False
)
```

index=False avoids saving index as a column unless you explicitly want to save the index

**2. Selecting Specific Columns While Reading**\
Imagine a CSV with 12 columns but you want only three

Bad:

```python
df = pd.read_csv("customers.csv")
```

Good:

```python
df = pd.read_csv(
    "customers.csv",
    usecols=[
        "First Name",
        "Last Name",
        "City"
    ]
)
```

```python
df
```

**Benefits:**\
Faster\
Less RAM\
Cleaner\
Professional almost always use usecols for large datasets.

**3. Choosing DataTypes While Reading**

```python
countries =  pd.read_csv(
    "customers.csv",
    usecols=["Country"],
    dtype={
        "Country":"category"
    }
)
```

```python
countries.info()
```

This is faster and more memory efficient than firstly loading a CSV and then changing the DataTypes of specific columns.

**4. Handling missing value during import**

Suppose your CSV contains:\
Salary\
50000\
NA\
Missing\
70000\
Tell Pandas these represent missing values

```python
df = pd.read_csv(
    "customers.csv",
    na_values=[
        "NA",
        "Missing"
    ]
)
```

Now, the missing values become NaN and can be dealt with easily.

**5. Reading Large CSVs**\
Usage of chunks is the best option

```python
chunks = pd.read_csv(
    "customers.csv",
    chunksize=10
)
```

```python
print(chunks)
```

```python
for chunk in chunks:
    print(chunk["City"])
```

This is exactly how ETL pipelines work.

**11. Common Parameters You Should Know**

Read first 100 rows

```python
pd.read_csv(
    "customers.csv",
    nrows=100
)
```

Skip first few rows

```python
pd.read_csv(
    "customers.csv",
    skiprows=5
)
```

Different delimiter

```python
pd.read_csv(
    "data.csv",
    sep=";"
)
```

No header

```python
pd.read_csv(
    "customers.csv",
    header=None
)
```

Custom Column names

```python
pd.read_csv(
    "data.csv",
    names=[
        "ID",
        "Name",
        "Salary"
    ]
)
```
