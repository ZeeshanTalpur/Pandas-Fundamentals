# <center> Data Cleaning Mastery
Before analysis, visualization, or machine learning, data must be cleaned.

**The Data Cleaning Workflow**

Whenever you load a dataset:

1. Inspect
2. Find Missing Values
3. Handle Missing Values
4. Find Duplicates
5. Remove Duplicates
6. Check Data Types
7. Fix Data Types
8. Clean Text Columns
9. Verify Results

This workflow should become automatic.

Let's create our dataset to work on

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "Name":["Ali","Sara","Ahmed","Ahmed","  Bilal","AYESHA"],
    "Age":[22,np.nan,30,30,35,28],
    "Salary":[50000,70000,np.nan,np.nan,120000,80000],
    "Department":["HR","IT","IT","IT","HR","Finance"]
})
print(df)
```

**1. Detect Missing Values**

```python
df.isna()
```

True --> missing\
False --> present

**Count Missing Values**

```python
df.isna().sum()
```

**Interpretation:**\
Age has 1 missing value\
Salary has 2 missing values

**2. Handling Missing Values**\
There is no universal solution.\
It depends on context.

**Method 1:** Drop rows\
Remove rows containing missing values.\
Only complete rows remain.\
Note: Should be used only when only a few rows are missing relative to the total.

```python
df.dropna()
```

Not beneficial in this case because, out of 5 rows, 2 contain missing values, removing which removes almost half the dataset.

**Method 2:** Fill with constant

```python
df.fillna("Unknown")
```

Useful for categorical columns

**Method 3:** Fill with mean\
Very common.

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
df
```

**Method 4:** Fill with Median\
preferred when outliers exist

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
df
```

**Method 5:** Fill with Mode\
Most frequent value

```python
df["Department"].mode()
```

```python
df["Department"].fillna(df["Department"].mode())
```

Useful for categorical columns

**3. Forward Fill:**\
Common in time series

```python
df.ffill()
```

**4. Backward Fill**

```python
df.bfill()
```

**5. Detect Duplicates**

```python
df.duplicated()
```

**Count Duplicates**

```python
df.duplicated().sum()
```

**Remove Duplicates**

```python
df = df.drop_duplicates()
```

```python
df
```

**6. DataType Inspection**

```python
df.dtypes
```

**7. Converting Data Types**

astype(): Convert to specific datatype

```python
df["Age"]=df["Age"].astype(int)
```

```python
df.dtypes
```

**Common Real-World Problem**\
CSV:\
Age\
22\
25\
30\
Unknown\
Column becomes:\
object\
instead of:\
int

```python
age = pd.Series([22,25,30,"unknown"])
print(f"Datatype due to unknown: {age.dtype}")
age = pd.to_numeric(age,errors="coerce")
print(f"Datatype after handling with pd.to_numeric():\n{age}")
```

Unknowns become NaN which can then be handled.

**8. Renaming Columns**

```python
df2 = pd.DataFrame({
    "cust_age":[12,16],
    "cust_sal":[20000,23000]
})
print(f"Before renaming: \n{df2}")
df2 = df2.rename(
    columns={
        "cust_age":"Age",
        "cust_sal":"Salary"
    }
)
print(f"After renaming: \n{df2}")
```

**9. Text Cleaning**

a) strip()\
remove extra whitespaces

```python
df["Name"] = df["Name"].str.strip()
df
```

b) lower()\
converts to lower-case

```python
df.loc[:,"Name"] = df.loc[:,"Name"].str.lower()
df
```

c) upper()\
converts to upper-case

```python
df.loc[:,"Name"] = df["Name"].str.upper()
df
```

d) title()\
converts to title case\
industry favorite

```python
df.loc[:,"Name"] = df["Name"].str.title()
df
```

e) replace\
replaces specific values in specific columns with other values

```python
df.loc[:,"Department"] = df["Department"].replace({
    "HR":"Human Resources",
    "IT":"Information Technology"
})
df
```
