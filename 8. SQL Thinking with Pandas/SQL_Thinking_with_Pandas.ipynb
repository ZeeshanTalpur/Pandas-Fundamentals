**The Big Idea:**\
SQL operates on TABLES\

**Dataset**

```python
import pandas as pd

df = pd.DataFrame({
    "Name":["Ali","Sara","Ahmed","Ayesha","Bilal"],
    "Age":[22,25,30,28,35],
    "Salary":[50000,70000,90000,80000,120000],
    "Department":["HR","IT","IT","Finance","HR"]
})
df
```

**1. SELECT**\
**SQL**\
SELECT Name, Salary\
FROM Employees;

**Pandas**

```python
df[["Name","Salary"]]
```

**2. WHERE**\
**SQL**\
SELECT *\
FROM Employees\
WHERE Salary > 80000;

**Pandas**

```python
df[df["Salary"]>80000]
```

Multiple Conditions:\
**SQL**\
SELECT * FROM Employees\
WHERE Department='IT'\
AND Salary > 70000;

**Pandas**


```python
df[
    (df["Department"]=="IT")
    &
    (df["Salary"]>70000)
    ]
```

**3. ORDER BY**\
**SQL**\
SELECT * FROM Employees\
ORDER BY Salary DESC;

**Pandas**

```python
df.sort_values(
    "Salary",
    ascending=False
)
```

Multiple Columns:\
**SQL**\
ORDER BY Department, Salary DESC

**Pandas**

```python
df.sort_values(
    ["Department","Salary"],
    ascending=[True,False]
)
```

**4. LIMIT**\

**6. GROUP BY**\
**SQL**\
SELECT Department,\
AVG(Salary)\
FROM Employees\
GROUP BY Department;

**Pandas**

```python
df.groupby("Department")["Salary"].mean()
```

Multiple Aggregations:\
**SQL**\
SELECT Department,\
AVG(Salary),\
MAX(Salary),\
COUNT(*)\
FROM Employees\
GROUP BY Department;

**Pandas**

```python
df.groupby("Department")["Salary"].agg("mean","max","count")
```

**7. HAVING**\
**SQL**\
SELECT Department,\
AVG(Salary)\
FROM Employees\
GROUP BY Department\
HAVING AVG(Salary) > 85000

**Pandas**

```python
grouped = df.groupby("Department")["Salary"].mean()
```

```python
grouped>85000
```

```python
grouped[grouped>85000]
```

**8. INNER JOIN**\
**SQL**\
SELECT *\
FROM Customers\
INNER JOIN Orders\
ON Customers.CustomerID\
=\
Orders.CustomerID

**Pandas**

```python
customers = pd.DataFrame({
    "CustomerID":[1,2,3],
    "Name":["Ali","Sara","Ahmed"]
})
orders = pd.DataFrame({
    "CustomerID":[1,2,1,4],
    "Product":["Laptop","Phone","Mouse","Keyboard"]
})
pd.merge(
    customers,
    orders,
    on="CustomerID"
)
```

**9. LEFT JOIN**\
**SQL**\
SELECT *\
FROM Customers\
LEFT JOIN Orders\
ON Customers.CustomerID\
=\
Orders.CustomerID

**Pandas**

```python
pd.merge(
    customers,
    orders,
    on="CustomerID",
    how="left"
)
```

**10. UNION ALL**\
**SQL**\
SELECT *\
FROM Employess2024\
UNION ALL\
SELECT *\
FROM Employees2025

**Pandas**

```python
df1 = pd.DataFrame({
    "Name":["Ali","Sara"],
    "Age":[23,25]
        })
df2 = pd.DataFrame({
    "Name":["Bilal","Shama"],
    "Age":[26,27]
})
pd.concat(
    [df1,df2],
    ignore_index=True
)
```

**11. CASE WHEN**\
**SQL**\
CASE \
WHEN Salary > 100000\
THEN 'Gold'\
ELSE 'Silver\
END

**Pandas**

```python
import numpy as np
df["Tier"] = np.where(
    df["Salary"] > 100000,
    "Gold",
    "Silver"
)
```

**15. SQL Query Translation**\
**SQL**\
SELECT Name, Salary\
FROM Employees\

ORDER BY Salary DESC\
LIMIT 2
