# <center>Exploratory Data Analysis

### What is EDA?

EDA means:

Understanding a dataset before modeling.

**The EDA Workflow**

A professional analyst often follows:

1. Overview
2. Descriptive Statistics
3. Categorical Analysis
4. Numerical Analysis
5. Correlation Analysis
6. Outlier Detection
7. Business Insights

This workflow should become automatic.

Our Dataset to work on:

```python
import pandas as pd
df = pd.DataFrame({
    "Name":["Ali","Sara","Ahmed","Ayesha","Bilal"],
    "Age":[22,25,30,28,35],
    "Salary":[50000,70000,90000,80000,120000],
    "Department":["HR","IT","IT","Finance","HR"],
})
df
```

**1. Describe**

```python
df.describe()
```

**Understanding Every Row**\
count : Number of non-missing values\
mean : Average\
std : Standard Deviation(Measures spread)\
min : Smallest value.\
25% : First Quartile (Q1)\
50% : Median (Q2)\
75% : Third Quartile (Q3)\
max : Largest value

**Stage 2: Categorical Analysis**

**10. Correlation Analysis**

```python
df[["Age","Salary"]].corr()
```

**Understanding Correlation Matrix**\
**Diagonal**\
Always : 1\
Because:\
Age vs Age\
Salary vs Salary\
Perfect correlation.\
Ignore diagonals.

**Positive Correlation:**\
+1 : Perfect positive relationship.\
**Negative Correlation:**\
-1 : Perfect negative relationship.\
**No linear relationship**\
0 : Zero Correlation

**Industry Interpretation:**\
0.8–1.0 : Strong Positive\
0.5–0.8 : Moderate Positive\
0–0.5 : Weak Positive

Negative values:

Same idea.

**12. Outlier Detection**

```python
q1 = df["Salary"].quantile(0.25)
q3 = df["Salary"].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 *iqr
outliers = df[
    (df["Salary"]<lower) | (df["Salary"]>upper)
    ]
```

**Professional EDA Checklist**
