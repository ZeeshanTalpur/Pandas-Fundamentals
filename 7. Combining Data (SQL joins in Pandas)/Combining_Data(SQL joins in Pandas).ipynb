# <center>Combining Data(SQL Joins in Pandas)

**The Big Picture**\
Multiple DataFrames --> Combine them --> Analyze Everything Together\
**The Three Major Tools**\
Pandas provides:\
pd.concat()\
pd.merge()\
df.join()

**1. pd.concat() - Stacking Data**\
Put things on top of each other\
OR\
Put things beside each other

**Vertical Concatenation (axis = 0)**

```python
import pandas as pd
df1 = pd.DataFrame({
    "Name":["Ali","Sara"],
    "Age":[23,25]
        })
df2 = pd.DataFrame({
    "Name":["Bilal","Shama"],
    "Age":[26,27]
})
combined = pd.concat([df1,df2])
combined
```

Notice: Indexes repitition is weird, isn't it?

**ignore_index=True**

```python
combined = pd.concat(
    [df1,df2],
    ignore_index=True
)
combined
```

Feels good now!

**Horizontal Concatenation(axis = 1)**

```python
names = pd.DataFrame({
    "Name":["Ali","Ahmed"]
})
ages = pd.DataFrame({
    "Age":[23,24]
})
salaries = pd.DataFrame({
    "Salary":[55000,75000]
})
df = pd.concat(
    [names,ages,salaries],
    axis=1
)
df
```

axis=0 --> stack rows --> Vertical\
axis=1 --> stack columns --> Horizontal

**3. pd.merge()**\
The Syntax:\
pd.merge(\
left_df,\
right_df,\
on="common_column"\
)

```python
customers = pd.DataFrame({
    "CustomerID":[1,2,3],
    "Name":["Ali","Sara","Ahmed"]
})
orders = pd.DataFrame({
    "CustomerID":[1,2,1,4],
    "Product":["Laptop","Phone","Mouse","Keyboard"]
})
```

```python
merged = pd.merge(
    customers,
    orders,
    on = "CustomerID"
)
merged
```

**Inner Join : Keep ONLY matching rows like above**

Next is:\
**Left Join : Keep all from the left df even if it deosn't exist in right**

```python
left = pd.merge(
    customers,
    orders, 
    on="CustomerID",
    how="left"
)
left
```

**Right Join: Keep all from the right df even if it doesn't exist in left**

```python
right = pd.merge(
    customers,
    orders,
    on = "CustomerID",
    how = "right"
)
right
```

**Outer Join : Keep EVERYTHING**

```python
outer = pd.merge(
    customers,
    orders,
    on = "CustomerID",
    how = "outer"
)
outer
```

**Dealing with different column names**

```python
customers = pd.DataFrame({
    "CustomerID":[1,2,3],
    "Name":["Ali","Sara","Ahmed"]
})
orders = pd.DataFrame({
    "CustID":[1,2,1,4],
    "Product":["Laptop","Phone","Mouse","Keyboard"]
})
```

In above example,\
customers --> CustomerID\
orders --> CustID

```python
merged = pd.merge(
    customers,
    orders,
    left_on = "CustomerID",
    right_on = "CustID"
)
merged
```

**Dealing with Multiple Keys**

```python
df1 = pd.DataFrame({
    "Name":["Ali","Sara"],
    "Age":[23,25],
    "Department":['HR','IT'],
    "Gender":["M","F"]
        })
df2 = pd.DataFrame({
    "Name":["Bilal","Shama"],
    "Age":[26,27],
    "Department":['Finance','IT'],
    "Gender":["M","F"]
})
```

**Updating Suffixes**

```python
multi_keys = pd.merge(
    df1,
    df2,
    on = ["Department","Gender"],
    how = "outer",
    suffixes = ("_left","_right")
)
multi_keys
```

**Indicator Column : useful for debugging joins**

```python
pd.merge(

    customers,

    orders,

    left_on = "CustomerID",
    right_on = "CustID",

    how="outer",

    indicator=True
)
```
