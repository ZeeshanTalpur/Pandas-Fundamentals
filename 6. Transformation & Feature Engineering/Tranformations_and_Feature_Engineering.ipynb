What is Feature Engineering?

Suppose we have:

Name	Age	Salary\
Ali	    22	50000\
Sara	35	120000

A business analyst might ask:

Create customer categories:

Salary < 60K  → Bronze\
60K-100K      → Silver\
100K+         → Gold

These categories don't exist.

We create them.

That's feature engineering.

**Dataset**

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "Name":["Ali","Sara","Ahmed","Ayesha","Bilal"],
    "Age":[22,25,30,28,35],
    "Salary":[50000,70000,90000,80000,120000],
    "Department":["HR","IT","IT","Finance","HR"]
})
df
```

**1. Creating New Columns**

```python
df["Bonus"] = df["Salary"]*0.1
```

```python
df
```

This is vectorized computation. No loops involved!

**2. map()**

```python
department={
    "HR" : "Human Resource"
}
df["Department"] = df["Department"].map(department)
```

```python
df
```

Notice that the missing keys become NaN using map()\
Let's create the dataset again and use another function that doesn't affect the unmatched keys.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    "Name":["Ali","Sara","Ahmed","Ayesha","Bilal"],
    "Age":[22,25,30,28,35],
    "Salary":[50000,70000,90000,80000,120000],
    "Department":["HR","IT","IT","Finance","HR"]
})
df["Bonus"] = df["Salary"]*0.1
df
```

That very function is discussed below:

**3. replace()**

```python
df["Department"] = df["Department"].replace({
    "HR" : "Human Resources"
}
)
df
```

Notice the unmatched keys don't become NaN. Replace() is more flexible whereas map() is faster for pure mapping.

```python
df["Department"] = df["Department"].replace({
    "IT" : "Information Technology"
}
)
df
```

**4. apply()**

used to apply a custom function to every value

```python
def age_category(age):
    if age<30:
        return "Young"
    return "Senior"
df["Category"] = df["Age"].apply(age_category)
```

```python
df
```

Lambda Functions : Same result, but compact logic

```python
df["Category"] = df["Age"].apply(
    lambda x: "Young" if x < 30 else "Senior"
)
```

```python
df
```

**When not to use apply()**\
If vectorized operation for something exist, don't use apply()\
In case, there's no vectorized operation for intended operation, use apply()

**5. np.where()**

```python
df["HighIncome"] = np.where(
    df["Salary"]>80000,
    "Yes",
    "No"
)
```

```python
df
```

Multiple Conditions:

```python
df["Level"] = np.where(
    df["Salary"]>100000,
    "Gold",
    np.where(
        df["Salary"]>70000,
        "Silver",
        "Bronze"
    )
)
df
```

**6. cut()**\
Used to categorize continuous data

```python
bins = [18,25,35,50]
labels = [
    "Young Adult",
    "Adult",
    "Middle Age",
]
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins,
    labels=labels
)
df
```

**Why Businesses Use This**
Instead of:\
Customer Age = 31

Businesses prefer:\
Customer Segment = Adult

Much easier to analyze.

**7. qcut()**\
Difference:\
cut() : Width of bins specified\
qcut() : Number of entries per bin specified

```python
df["SalaryLevel"] = pd.qcut(
    df["Salary"],
    q=3,
    labels=["Low","Medium","High"]
)
df
```

Very useful for:\
Customer segmentation\
Risk levels\
Marketing campaigns

**8. assign()**

```python
df = df.assign(
    Bonus = df["Salary"]*0.1,
    Tax = df["Salary"]*0.2
)
df
```
