## <center> Series and DataFrames Fundamentals</center>

**1. Series**\
**Think:**\
Series = One Column\
**Equivalent to:**\
NumPy 1D Array\
+\
Labels

```python
import pandas as pd
ages = pd.Series([22,25,30])
ages
```

Notice that indices also appear along with the series, and the datatype of the series too

**Anatomy of Series**\
Series contains\
a) Values

```python
ages.values
```

b) Index


```python
ages.index
```

c) DataType

```python
ages.dtype
```

d) Custom Index: Indexes do not necessarily have to be numbers

```python
ages = pd.Series(
    [22,25,30],
    index = ['Ali','Sara','Ahmed']
)
ages
```

```python
ages['Sara']
```

This is the foundation of Pandas.

**2) DataFrames:**
**Think:**\
DataFrame = Multiple Series Together

```python
df = pd.DataFrame({
    "Name": ["Ali","Sara","Ahmed"],
    "Age": [22,25,30],
    "Salary": [50000,70000,90000]
})
df
```

**Anatomy of a DataFrame**\
Every DataFrame contains:\
Rows\
Columns\
Index\
Values\
Datatypes

**Ways to Create DataFrames**\
a) DataFrame from a NumPy Array:

```python
data = np.array([
    [22,50000],
    [25,70000],
    [30,90000]
])
df = pd.DataFrame(
    data,
    columns=["Age","Salary"]
)
df
```

b) DataFrame from a Dictionary: (Most Common)

```python
df = pd.DataFrame({
    "Age":[22,25,30],
    "Salary":[50000,70000,90000]
})
df
```

c) DataFrame from a list of dictionaries: (Very common in APIs)

```python
df = pd.DataFrame([
    {"Age":22,"Salary":50000},
    {"Age":25,"Salary":70000},
    {"Age":30,"Salary":90000}
])
df
```

**Essential Attributes**\
a) Shape

```python
df.shape
```

b) Columns

```python
df.columns
```

c) Index

```python
df.index
```

d) dtypes

```python
df.dtypes
```

e) ndim

```python
df.ndim
```

f) values

```python
df.values
```

g) size

```python
df.size
```

**info()**\
One of the most important Pandas functions.

```python
df.info()
```

Industry analysts run this almost immediately after loading data.

**describe()**\
Quick statistical summary.

```python
df.describe()
```
