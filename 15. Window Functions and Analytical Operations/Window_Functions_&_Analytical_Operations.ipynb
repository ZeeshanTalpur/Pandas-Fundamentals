# <center>Window Functions and Analytical Operations

Window Functions answer "Tell me something about each row using information from surrounding rows"\
This is exactly how analysts, financial engineers and data scientists analyze trends.\
**What are window functions?**\
Imagine daily sales:\
Day 1 --> 100\
Day 2 --> 120\
Day 3 --> 150\
Day 4 --> 90\
Day 5 --> 180\
Instead of asking: "Total sales?"\
Ask : \
"Average of last 3 days?"\
"Difference from yesterday?"\
"Growth since yesterday?"\
"Running total?"\
"Rank among employees?"\
"Highest sale so far?"\
These are all window operations

**Dataset**

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Date":pd.date_range(
        "2025-01-01",
        periods=10
    ),
    "Sales":[100,120,150,130,160,180,170,210,220,200],
    "Employee":["A","A","A","A","A","B","B","B","B","B"]
})

df
```

**1. shift()**\
Probably the most used window function\
It moves values without changing the index\
Current row can "look" at previous or next rows

```python
df["Sales"].shift(1)
```

Previous Day Sales:

```python
df["Yesterday"] = df["Sales"].shift(1)
df
```

Next day sales:

```python
df["Tomorrow"] = df["Sales"].shift(-1)
df
```

**2. diff()**\
Difference between consecutive rows\
Formula : Current - Previous

```python
df["Daily Change"] = df["Sales"].diff()
df
```

**3. pct_change()**\
Percentage growth\
Formula: (Current - Pevious) / Previous

```python
df["Growth"] = df["Sales"].pct_change()
df
```

Multiply by 100 if desired

```python
df["Growth %"] = (
    df["Sales"]
    .pct_change()
    *100
)
df
```

**4. cumsum()**\
Running Total

```python
df["Running Sales"] = df["Sales"].cumsum()
df
```

**5. cummax()**\
Maximum value so far

```python
df["Highest"] = df["Sales"].cummax()
```

```python
df
```

**6. cummin()**\
Lowest value encountered

```python
df["Lowest"] = df["Sales"].cummin()
df
```

**7. cumprod()**\
Running Multiplication

```python
s = pd.Series([2,3,4])
s.cumprod()
```

**8. rank()**\
Rank rows

```python
df["Rank"] = df["Sales"].rank(ascending=False)
df
```

Methods:

```python
df["Rank_Avg"] = df["Sales"].rank(method="average")
df["Rank_Dense"] = df["Sales"].rank(method="dense")
df["Rank_first"] = df["Sales"].rank(method="first")
df
```

**9. rolling()**\
The king of window functions

Three Day Moving Average

```python
df["Rolling Mean"] = (
    df["Sales"]
    .rolling(window=3)
    .mean()
)
df
```

Rolling sum

```python
df["Rolling Sum"] = (
    df["Sales"]
    .rolling(window=3)
    .sum()
)
df
```

Rolling min

```python
df["Rolling Minimum"] = (
    df["Sales"]
    .rolling(window=3)
    .min()
)
df
```

Rolling Max

```python
df["Rolling Max"] = (
    df["Sales"]
    .rolling(window=3)
    .max()
)
df
```

Rolling Standard Deviation

```python
df["Rolling Std"] = (
    df["Sales"]
    .rolling(window=3)
    .std()
)
df
```

**10. expanding()**\
Unlike rolling, it never forgets previous rows\
Happens to behave like cumulative functions

```python
df["Expanding Mean"] = (
    df["Sales"]
    .expanding()
    .mean()
)
df
```

**11. ewm()**\
Exponentially Weighted Moving Average

```python
df["EWMA"] = (
    df["Sales"]
    .ewm(alpha=0.3)
    .mean()
)
df
```

**12. Group Wise Window Functions**\
Instead of one running total, Compute per employee

```python
df["Employee Running Sales"] = (
    df
    .groupby("Employee")["Sales"]
    .cumsum()
)
df
```
